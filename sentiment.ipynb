```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
entries = [
    {"date": "2025-06-25", "text": "I had such a lovely day at the beach."},
    {"date": "2025-06-26", "text": "Feeling exhausted and overwhelmed with work."},
    {"date": "2025-06-27", "text": "Today was pretty uneventful."},
    {"date": "2025-06-28", "text": "I’m really excited about tomorrow’s trip!"},
    {"date": "2025-06-29", "text": "Nothing went right today. It was terrible."}
]
```

```python
df = pd.DataFrame(entries)
df['date'] = pd.to_datetime(df['date'])
```

```python
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['sentiment'] = df['text'].apply(get_sentiment)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['date'], df['sentiment'], marker='o', linestyle='-', color='skyblue')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Mood Over Time Based on Journal Entries', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Sentiment Polarity (−1 = Negative, +1 = Positive)')
plt.grid(True)
plt.show()
```
